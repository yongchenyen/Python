讀取資料

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/2018年_倫敦天氣資料.csv')
print(df)
```

納入座標資料，把Tmax曲線取出

```python
df.plot(x='Month',y='Tmax')
```

列入網格狀態：grid

```python
df.plot(x='Month',y='Tmax',grid=True)
```

將最高溫與最低溫同步查詢

```python
df.plot(x='Month',y=['Tmax','Tmin'])
```

建立bar線條圖示說明

```python
df.plot(x='Month',y=['Tmax','Tmin'],kind='bar')
```

同時將不同欄位進行顯示

```python
df.plot(
    kind='line', y=['Tmax','Tmin','Rain','Sun'], #4個變數可視化
    subplots=True, #多子圖併存
    layout=(2,2), #子圖排列2行2列
    figsize=(20,10) #圖布的尺寸
    )
```
